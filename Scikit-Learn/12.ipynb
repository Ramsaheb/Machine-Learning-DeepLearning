```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train
```

```python
df = pd.concat([train, test], sort=False)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.drop('Cabin', axis=1)
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='mean')
df['Age'] = si.fit_transform(df[['Age']])
df['Survived'] = si.fit_transform(df[['Survived']])
df['Fare'] = si.fit_transform(df[['Fare']])
df.isnull().sum()
```

```python
df.info()
```

```python
si2 = SimpleImputer(strategy='most_frequent')
df['Embarked'] = si2.fit_transform(df[['Survived']])
df.isnull().sum()
```

```python
df = df.drop(['Name', 'Ticket'], axis=1)
```

```python
df.head()
```

```python
threshold = 3
from scipy.stats import zscore

numerical_features = ['Fare', 'Age']
z_scores = np.abs(zscore(df[numerical_features]))
```

```python
z_scores
```

```python
outliers = np.where(z_scores > threshold)
outliers
```

```python
from scipy.stats.mstats import winsorize

df['Age'] = winsorize(df['Age'], limits=[0.15, 0.15])
df['Fare'] = winsorize(df['Fare'], limits=[0.15, 0.15])
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()

df['Sex'] = oe.fit_transform(df[['Sex']])
df['Embarked'] = oe.fit_transform(df[['Embarked']])
df['Survived'] = oe.fit_transform(df[['Survived']])
```

```python
df.head(2)
```

```python
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ab = AdaBoostClassifier(n_estimators=45, learning_rate=1, random_state=0)

ab.fit(X_train, y_train)
y_pred = ab.predict(X_test)
```
